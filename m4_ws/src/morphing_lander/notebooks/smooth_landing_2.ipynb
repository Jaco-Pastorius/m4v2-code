```python
import numpy as np
import matplotlib.pyplot as plt

# parameters
g    = 9.81 
kT   = 28.0
Tmax = 4*kT
vmax = np.pi/8
m    = 5.5

# get landed values
p    = 10.0
l    = 0.26
phistar = np.arctan(p/l)
zstar   = -0.11

# landing height for phi=90 degrees
zland = zstar + l*(1-np.sin(phistar))

# phi0 as a function of z0
phi0 = lambda z0 : np.pi/2 - vmax/g*np.sqrt(-2*g*(z0-zland))
zdot = lambda phi0 : g*(np.pi/2-phi0)/vmax

z0 = np.linspace(-2.0,zland,200)
plt.plot(z0,np.rad2deg(phi0(z0)))
# plot zland
plt.plot(zstar,np.rad2deg(phistar),'ro')
plt.xlabel('z0')
plt.ylabel('phi0')
plt.title('max angle for a given z0')
plt.show()

# plot zdot
plt.plot(z0,zdot(phi0(z0)))
# plot 1 on the graph
plt.plot(z0,np.ones(z0.shape),'r--')

plt.xlabel('z0')
plt.ylabel('zdot')
plt.title('zdot for a given z0')
plt.show()



```
